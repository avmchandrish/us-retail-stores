```python
#creating a function to be called for each source of Albertsons
def albertsons_companies_store(store):
    #Data Frames
    col=['source', 'title', 'address_line', 'city', 'postal_code', 'ph_no', 'state_name', 'state_code', 'country']
    df_final=pd.DataFrame(columns=col)

    sitemap=requests.get('https://local.' + str(store) + '.com/sitemap.xml')  #store locator sitemap
    sm_soup=BeautifulSoup(sitemap.text, 'xml')                              #parsing it with a xml parser
    for link in tqdm(sm_soup.find_all('xhtml:link')):                          #loop through all the store site links present
        if link.get('href').count('/')==5:
            url=link.get('href')
            loc_site=requests.get(url)
            soup_loc_site=BeautifulSoup(loc_site.text)
            source=soup_loc_site.find_all('div', class_='Heading--sub ContentBanner-subtitle')[0].text
            title=soup_loc_site.find_all('div', class_='Heading--lead ContentBanner-title')[0].text
            address_line=soup_loc_site.find_all('span', class_="c-address-street-1")[0].text
            city=soup_loc_site.find_all('span', class_="c-address-city")[0].text
            postal_code=soup_loc_site.find_all('span', class_="c-address-postal-code")[0].text
            ph_no=soup_loc_site.find_all('a', class_="Phone-link")[0].text
            state_name=soup_loc_site.find_all('abbr')[0].get('title')
            state_code=soup_loc_site.find_all('abbr')[0].text
            country=soup_loc_site.find_all('abbr')[1].get('title')

            row=[source, title, address_line, city, postal_code, ph_no, state_name, state_code, country]
            df_temp=pd.DataFrame([row], columns=col)
            df_final=df_final.append(df_temp)

    df_final.to_csv(str(store) + '_stores.csv')
```

```python
store_list=['safeway', 'vons', 'jewelosco', 'shaws']
for store in store_list:
    albertsons_companies_store(store)

```
